**Import Dependencies**

```python
import json
from matplotlib import pyplot as plt
```

**Load Training History For Each Model**

```python
SAVE_PATH = 'gdrive/MyDrive/Sentiment_Analysis/'

# Load model variations training history
single_stemmed = json.load(open(SAVE_PATH + 'LSTM_single_stemmed_training.json', 'r'))
stacked_stemmed = json.load(open(SAVE_PATH + 'LSTM_stacked_stemmed_training.json', 'r'))
stacked_unstemmed = json.load(open(SAVE_PATH + 'LSTM_stacked_unstemmed_training.json', 'r'))
```

**Plot Training Testing Loss**

```python
# Plot testing loss history (solid line)
plt.plot(single_stemmed['val_loss'][:5], 'r')      
plt.plot(stacked_unstemmed['val_loss'], 'b')
plt.plot(stacked_stemmed['val_loss'], 'g')

# Plot training loss history (dashed line)
plt.plot(single_stemmed['loss'][:5], 'r', linestyle='dashed')      
plt.plot(stacked_unstemmed['loss'], 'b', linestyle='dashed')
plt.plot(stacked_stemmed['loss'], 'g', linestyle='dashed')

plt.title('Model Loss History Comparison')
plt.ylabel('Loss: Train (dashed) / Test (solid)')
plt.xlabel('Epoch')
plt.legend(['Single LSTM, Stemmed Data', 'Stacked LSTM, Unstemmed Data', 'Stacked LSTM, Stemmed Data'])
plt.savefig(SAVE_PATH + 'figures/loss_plot_comparison.png')
plt.show()
```

**Plot Testing Accuracy**

```python
# Plot testing accuracy history (solid line)
plt.plot(single_stemmed['val_accuracy'][:5], 'r')      
plt.plot(stacked_unstemmed['val_accuracy'], 'b')
plt.plot(stacked_stemmed['val_accuracy'], 'g')

# Plot training accuracy history (dashed line)
plt.plot(single_stemmed['accuracy'][:5], 'r', linestyle='dashed')      
plt.plot(stacked_unstemmed['accuracy'], 'b', linestyle='dashed')
plt.plot(stacked_stemmed['accuracy'], 'g', linestyle='dashed')

plt.title('Model Accuracy History Comparison')
plt.ylabel('Accuracy: Train (dashed) / Test (solid)')
plt.xlabel('Epoch')
plt.legend(['Single LSTM, Stemmed Data', 'Stacked LSTM, Unstemmed Data', 'Stacked LSTM, Stemmed Data'])
plt.savefig(SAVE_PATH + 'figures/acc_plot_comparison.png')
plt.show()
```
